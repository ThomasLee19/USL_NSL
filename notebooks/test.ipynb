```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
import matplotlib.pyplot as plt

# Test Isolation Forest
print("Testing Isolation Forest...")
iso_forest = IsolationForest(random_state=42)
X = np.random.rand(100, 2)  # Random data for testing
iso_forest.fit(X)
predictions = iso_forest.predict(X)
print("Isolation Forest predictions:", predictions[:10])  # Print first 10 predictions

# Test One-Class SVM
print("\nTesting One-Class SVM...")
one_class_svm = OneClassSVM()
one_class_svm.fit(X)
predictions = one_class_svm.predict(X)
print("One-Class SVM predictions:", predictions[:10])  # Print first 10 predictions

# Test plotting (optional)
plt.scatter(X[:, 0], X[:, 1], c=predictions, cmap='viridis')
plt.title("Test Plot for Environment Check")
plt.show()

print("\nAll libraries are imported successfully, and env is configured correctly!")
```
